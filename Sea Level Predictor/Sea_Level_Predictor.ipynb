```python
# libraries
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
```

```python
# Load the data
url = 'https://raw.githubusercontent.com/404reese/ML-projects/main/Sea%20Level%20Predictor/epa-sea-level.csv'
df = pd.read_csv(url)
```

```python
# Create a scatter plot
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
# Get the slope and y-intercept of the line of best fit
slope, intercept, r, p, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
# Plot the line of best fit
years = np.array([min(df['Year']), max(df['Year']), 2050])
plt.plot(years, slope*years + intercept, 'r', label='All data')
```

```python
# Get the slope and y-intercept of the line of best fit for data from 2000 onwards
df_recent = df[df['Year'] >= 2000]
slope_recent, intercept_recent, r, p, std_err = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])
```

```python
# Plot the line of best fit for data from 2000 onwards
plt.plot(years, slope_recent*years + intercept_recent, 'g', label='Data from 2000')
```

```python
# Add labels and title
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
```

```python
# Add legend
plt.legend()
```

```python
# Save and return the image
plt.savefig('sea_level_rise.png')
plt.show()
```
